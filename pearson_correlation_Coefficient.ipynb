```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
data=pd.read_csv('/content/sample_data/Salary_Data.csv')
data.head()
```

know the last five elements,information about data and description

```python
data.tail()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(12,6))
sns.pairplot(data,x_vars=['YearsExperience'],y_vars=['Salary'],size=7,kind='scatter')
plt.xlabel('Years')
plt.ylabel('salary')
plt.title('Salary Predection')
plt.show()
```

Cooking data

```python
X=data.iloc[:,:-1]
y=data.iloc[:,1]
```

```python
X.head()
```

```python
y.head()
```

```python
X_Train,X_Test,y_Train,y_Test=train_test_split(X,y,train_size=.8,random_state=10)
```

```python
model=LinearRegression()
model.fit(X_Train,y_Train)
```

```python
plt.scatter(X_Train,y_Train,color='yellow')
plt.plot(X_Train,model.predict(X_Train),color='red')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
y_pred=model.predict(X_Test)
plt.scatter(X_Test,y_Test,color='yellow')
plt.plot(X_Test,y_pred,color='red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
c=[i for i in range(1,len(y_Test)+1)]
plt.plot(c,y_Test,color='red',linestyle='-')
plt.plot(c,y_pred,color='yellow',linestyle='-')
plt.xlabel('index')
plt.ylabel('salary')
plt.title('Prediction')
plt.show()
```

Plotting the Error


```python
c=[i for i in range(1,len(y_Test)+1)]
plt.plot(c,y_Test-y_pred,color='red',linestyle='-')
plt.xlabel('index')
plt.ylabel('error')
plt.title('Error Value')
plt.show()
```

```python
mse=mean_squared_error(y_Test,y_pred)
r2=r2_score(y_Test,y_pred)
print("Mean Squared Error",mse)
print("R2 Score",r2)
```

Intecept and coeff of the line

```python
print('Intercept of the model:',model.intercept_)
print('Coefficient of the line:',model.coef_)
```
